```python
import pandas as pd
import re

keyword = '칼부림'
df_r = pd.read_csv(f'{keyword}_naver_news.csv')
df_r
```

```python
def extract_word(text):
    hangul = re.compile('[^가-힣]') 
    result = hangul.sub(' ', text) 
    return result

# 한글이 아닌 형태소 제거
df_r['content'] = df_r['content'].apply(lambda x:extract_word(x))
df_r['title'] = df_r['title'].apply(lambda x:extract_word(x))

# dataframe 에서 list로 자료형 변환
title_list = df_r.title.values.tolist()
content_list = df_r.content.values.tolist()

print(type(content_list), len(content_list))
```

```python
import nltk
from konlpy.tag import Okt

t = Okt()

# 형태소 분석기 input 타입에 맞춰 string으로 변환
content_text = ''

for each_line in title_list:
    content_text = content_text + each_line
    
for each_line in content_list:
    content_text = content_text + each_line
    
content_text
```

```python
# 명사 추출
tokens_ko = t.morphs(content_text)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko)   
print(len(ko.tokens))          # 토큰 전체 개수
print(len(set(ko.tokens)))     # 토큰 unique 개수
```

```python
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(100)    # 가장 많이 나온 단어 100개 선정
```

```python
# 불용어 : 인터넷 검색 시 검색 용어로 사용하지 않는 단어. 관사, 전치사, 조사, 접속사 등 검색 색인 단어로 의미가 없는 단어
stop_words = [')','?','1','1~2','"(', '_', ')/','\n','\n\n','\n\n\n','.',',', '<','!','(','(', '??','..', '4', '|', '>', '?(', '"…', '#', '&', '・', "']",'.',' ','/',"'",'’','”','“','·', '[','!','\n','·','‘','"','\n ',']',':','…',')','(','-', 'nan','가','요','답변','...','을','수','에','질문','제','를','이','도',
              '좋','1','는','로','으로','2','것','은','다',',','니다','대','들',
              '2017','들','데','..','의','때','겠','고','게','네요','한','일','할',
              '10','?','하는','06','주','려고','인데','거','좀','는데','~','ㅎㅎ',
              '하나','이상','20','뭐','까','있는','잘','습니다','다면','했','주려',
              '지','있','못','후','중','줄','6','과','어떤','기본','!!','…\'',
              '단어','라고','중요한','합','가요','....','보이','네','무지','했습니다',
              '이다','대해','에게','입니다','있다','대한','3','합니다','및','장','에서','하고','검','한다','만',
              '적', '성', '삼', '등', '전','됐다','인', '그', '했다', '와', '위', '해', '권', '된', '서', '말', '분', '칼부림']

# 불용어 제거
tokens_ko = [each_word for each_word in tokens_ko
             if each_word not in stop_words]
print(len(tokens_ko))

# 한 글자 제거
tokens_ko = [x for x in tokens_ko if len(x)>1]
print(len(tokens_ko))

# 토큰 빈도수 확인
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(20000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 그래프에서 한글 폰트 깨지는 문제에 대한 대처(전역 글꼴 설정)
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)

plt.figure(figsize=(16,8))
ko.plot(50)
plt.show()
```

```python
from wordcloud import WordCloud
from PIL import Image

data = ko.vocab().most_common(300)

print(len(data))
data
```

```python
# list를 딕셔너리로 만들어주는 함수
def todict(list_tuple):
    todict = {}
    for i in range(0, len(list_tuple)):
        todict[data[i][0]] = data[i][1]
    return todict

wordcloud = WordCloud(font_path = 'c:/Windows/Fonts/malgun.ttf',
                      relative_scaling = 0.2,
                      background_color = 'white',
                     ).generate_from_frequencies(todict(data))

plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
